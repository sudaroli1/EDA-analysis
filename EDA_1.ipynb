<h3><font color = 'orange'>From the data obtained with the missing values count different columns have different count for the null values so let us perform the filling up of the null values.</h3>
<h4><font color = 'purple'> since the LB represents the average fetel rate over a period, let us fill those null values with the mean as this is a numerical value</h4>

```python
updated_cardiac = cardiac_dataframe
updated_cardiac['LB']=updated_cardiac['LB'].fillna(updated_cardiac['LB'].mean())
updated_cardiac['AC']=updated_cardiac['AC'].fillna(updated_cardiac['AC'].mean())
updated_cardiac['DS']=updated_cardiac['DS'].fillna(updated_cardiac['DS'].mean())
updated_cardiac['DP']=updated_cardiac['DP'].fillna(updated_cardiac['DP'].mean())
updated_cardiac['MLTV']=updated_cardiac['MLTV'].fillna(updated_cardiac['MLTV'].mean())
updated_cardiac['Width']=updated_cardiac['Width'].fillna(updated_cardiac['Width'].mean())
updated_cardiac['Tendency']=updated_cardiac['Tendency'].fillna(updated_cardiac['Tendency'].mean())
updated_cardiac['NSP']=updated_cardiac['NSP'].fillna(updated_cardiac['NSP'].mean())
updated_cardiac.info()
```

```python
cardiac_dataframe.isnull().sum()
```

<h4><font color = 'purple'> since the other columns with missing values also holds numerical values the imputation is performed to replace them with mean values.</font></h4>
<h5><font color = 'black'> As seen the missing values imputation is performed for all the missing values columns</font></h5>
<h5><font color = 'red'>Since all the data types represents the float data type inconsistencies in the form of numbers been stored as strings is not present</h5>

<h1>What are outliers? How to detect them?</h1>
<h2><font color= 'red'>Outliers are basically instances of the data that possess high difference among the other data instances. So this is created because of mistake or variance.</h2>
<h3><font color= 'green></h3>

```python
# Identifying the Outliers
import seaborn as sns
from matplotlib import pyplot as plt
plt.figure(figsize=(15,8))
sns.boxplot(x="variable", y="value", data=pd.melt(updated_cardiac))
```

<h1><font color='blue'> As observed from the plotting, the variables LB,ASTV,MSTV,ALTV,MLTV and width possess outliers</h1>
<h2><font color='pink'> These outliers are treated using the IQR method</h1>

```python
#Treating the outliers using IQR
Q1 = updated_cardiac.quantile(0.25)
Q3 = updated_cardiac.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
updated_cardiac_out = updated_cardiac[~((updated_cardiac < (Q1 - 1.5 * IQR)) |(updated_cardiac > (Q3 + 1.5 * IQR))).any(axis=1)]
updated_cardiac_out.shape
```

```python
# plotting the box after the outliers are identified and treated
import seaborn as sns
from matplotlib import pyplot as plt
plt.figure(figsize=(15,8))
sns.boxplot(x="variable", y="value", data=pd.melt(updated_cardiac_out))
```

<h2><font color='purple'> As observed from the above figure the outliers present in the LB,ASTV,MSTV,ALTV,MLTV and width are handled </h1>
